# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Barplot and Count Plot

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'Set1')
```

```python
sns.barplot(data = tips, palette = 'Set1', orient = 'h')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, palette = 'coolwarm')
```

```python
tips['percent'] = (tips['tip']/tips['total_bill'])
```

```python
tips.head(1)
```

```python
sns.boxplot(x = 'day', y = 'percent', hue = 'sex', data = tips, palette = 'coolwarm')
```

# Violinplot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips,hue = 'sex', split = True, pallet = 'rainbow')
```

# Stripplot and Swarmplots

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, palette = 'Set1', split = True)
```

```python
sns.violinplot(x = 'tip', y = 'day', data = tips, palette = 'rainbow')
sns.swarmplot(x = 'tip', y = 'day', data = tips, color = 'black', size = 3)
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tips, palette = 'rainbow')
sns.swarmplot(x = 'day', y = 'tip', data = tips, color = 'black', size = 3)
```
